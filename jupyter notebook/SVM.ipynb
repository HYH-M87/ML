```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
```

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
X , y  = datasets.make_moons( noise=0.15 , random_state=666 )

plt.scatter(X[y==0 , 0] , X[y==0 , 1])
plt.scatter(X[y==1 , 0] , X[y==1 , 1])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

def RBFKernelSVC(gamma = 1.0):
    return Pipeline([
        ("stand_sca" , StandardScaler()),
        ("svc" , SVC(kernel = 'rbf' , gamma = gamma)),
    ])
```

```python
svc_gamma1 = RBFKernelSVC(gamma = 1.0)
svc_gamma1.fit(X , y)
```

```python
plot_decision_boundary(svc_gamma1 , [-1.5 , 2.5 , -1.0 , 1.5])
plt.scatter(X[y==0 , 0] , X[y==0 , 1])
plt.scatter(X[y==1 , 0] , X[y==1 , 1])
plt.show()
```

```python
svc_gamma10 = RBFKernelSVC(gamma = 10)
svc_gamma10.fit(X , y)
```

```python
plot_decision_boundary(svc_gamma10 , [-1.5 , 2.5 , -1.0 , 1.5])
plt.scatter(X[y==0 , 0] , X[y==0 , 1])
plt.scatter(X[y==1 , 0] , X[y==1 , 1])
plt.show()
```

## SVM 解决回归问题

和解决分类相反，其思路是，找到一个ε，使其包含的数据点最多，并最小化这个ε

```python
boston = datasets.load_boston()
X = boston.data
y = boston.target
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y)
```

```python
from sklearn.svm import LinearSVR
from sklearn.svm import SVR

def StanderLinearSVR(epsilon = 0.1):
    return Pipeline([
        ("stander_sca" , StandardScaler()),
        ("svr" , LinearSVR(epsilon = epsilon))
    ])
```

```python
svr = StanderLinearSVR(epsilon = 0.1)
svr.fit(X_train , y_train)
```

```python
svr.score(X_test , y_test)
```
